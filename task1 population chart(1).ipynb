```python
import pandas as pd

# Replace the filename below with your actual uploaded CSV filename
df = pd.read_csv("API_SP.POP.TOTL_DS2_en_csv_v2_124794.csv", skiprows=4)

# Show the first few rows of the dataset
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Choose a few countries to display
selected_countries = ['India', 'China', 'United States', 'Indonesia', 'Brazil']

# Filter data for those countries and 2022
data = df[df['Country Name'].isin(selected_countries)][['Country Name', '2022']]

# Plot bar chart
plt.figure(figsize=(10,6))
sns.barplot(x='Country Name', y='2022', data=data, palette='mako')
plt.title('Population of Selected Countries (2022)')
plt.xlabel('Country')
plt.ylabel('Population')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Drop rows with missing values in 2022
df_clean = df[['Country Name', '2022']].dropna()

# Plot histogram
plt.figure(figsize=(10,6))
sns.histplot(df_clean['2022'], bins=30, kde=True)
plt.title('Distribution of Population Across Countries (2022)')
plt.xlabel('Population')
plt.ylabel('Number of Countries')
plt.show()
```
